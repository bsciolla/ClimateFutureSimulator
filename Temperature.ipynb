```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
coef1 = (2.9-1.22)/(1225-500)/3.7;
const1 = 1.22 - 500*3.7 * coef1;
print(coef1, const1)
```

```python
coef2 = (4.45-2.9)/(2000-1225)/3.7;
const2 = 2.9 - 1225*3.7 * coef2;
print(coef2, const2)
```

```python
def temp(C):
    if C > (1225*3.7):
        return const2 + coef2 * C;
    else:
        return const1 + coef1 * C;
vtemp = np.vectorize(temp)
```

```python
x = np.linspace(start=0, stop=10000, num=50)
y = vtemp(x)
```

```python
plt.plot(x, y)
plt.show()
```
